```python
## In what year was food produced the most

c = f_prod.groupby("Year")[['Value']].sum().reset_index()

fig, ax = plt.subplots(figsize=(12, 7) )
hbars = ax.bar(c.Year, c.Value, align='center',width =0.5)
ax.bar_label(hbars, labels=c.Value,
                 rotation=90, padding=10, fontsize=12.5)
ax.set (yticklabels= []);
plt.xlabel('Year', fontsize=15)
plt.ylabel('Food Produced (kilotons)', fontsize=15)
plt.title('Total Food Produced For Each Year ', fontsize=15);
```

```python
total_prod.rename(columns={'Value': 'value_produced'}, inplace=True)
total_prod.rename(columns={'Supplied_value': 'value_supplied'}, inplace=True)
```

```python
food = total_prod.copy()
```

```python
food
```

```python
d = food.groupby("Country")[['value_produced']].sum().reset_index()
```

```python
y = food.groupby("Country")[['value_produced','value_supplied']].sum().reset_index()
```

```python
u = y.sort_values(by='value_produced')
v= y.sort_values(by='value_supplied')
```

```python
d.plot(x="Country", y=["value_produced"],
       kind="bar", width = 0.8, figsize = (15,10));
plt.title('Average budget, revenue and profit over the years by genres '.title(),
         fontsize=15)
plt.xlabel('Genre', fontsize=15)
plt.ylabel('Average', fontsize=15);
plt.xticks( fontsize=12);
```

```python
y.plot(x="Country", y=["value_produced", 'value_supplied'],
       kind="bar", width = 0.8, figsize = (15,10));
plt.title('Average budget, revenue and profit over the years by genres '.title(),
         fontsize=15)
plt.xlabel('food Production and Supply', fontsize=15)
plt.ylabel('Quatity', fontsize=15);
plt.xticks( fontsize=12);




```

```python
u.plot(x="Country", y=[ 'value_produced'],
       kind="barh", width = 0.8, figsize = (13,10));
plt.title('Total Food production by each countries from 2004-2013'.title(),
         fontsize=15)
plt.xlabel('Total Value Produced (kilotons)', fontsize=15)
plt.ylabel('Countries', fontsize=15);
plt.xticks( fontsize=12);
plt.bar_label(hbars, labels=u.value_produced,
                 rotation=0, padding=10, fontsize=12.5);

plt.legend(loc='lower right'); 
```

```python
 
v.plot(x="Country", y=[ 'value_supplied'],
       kind="barh", width = 0.8, figsize = (13,10));
plt.title('Total Food supplied by each countries from 2004-2013'.title(),
         fontsize=15)
plt.xlabel('Total Value Supplied (kcal/person/day)', fontsize=15)
plt.ylabel('Countries', fontsize=15);
plt.xticks( fontsize=12);


plt.legend(loc='lower right'); 
```
